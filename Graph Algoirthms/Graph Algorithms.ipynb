```python
import networkx as nx
import networkx.algorithms.tree as tree
from networkx.generators.random_graphs import erdos_renyi_graph
```

```python

G0 = nx.Graph()

G0.add_edge(1, 2) # default edge data=1
G0.add_edge(2, 3)
G0.add_edge(1, 3)
G0.add_edge(1, 4)
G0.add_edge(4, 5)
G0.add_edge(4, 6)

nx.draw(G0, with_labels = True, node_color='lightgreen')
```

```python
G = nx.Graph()
elist = [(1, 2), (2, 3), (1,3),(1, 4), (4, 2), (4, 6), (5, 6), (5,7), (5,8), (6,7) ]
#nlist = nx.Graph([(0, 1), (1, 2), (2, 0)]) 
G.add_edges_from(elist)
#G.add_nodes_from(nlist)
#elist = [('a', 'b', 5.0), ('b', 'c', 3.0), ('a', 'c', 1.0), ('c', 'd', 7.3)]
#G.add_weighted_edges_from(elist)

nx.draw(G,  with_labels = True, node_color='yellow')
```

```python
# Prim Algorithm
mst = tree.minimum_spanning_edges(G, algorithm='prim', data=False)
edgelist = list(mst)
sorted(edgelist)

G2 = nx.Graph()
G2.add_edges_from(edgelist)

nx.draw(G2,  with_labels = True, node_color='yellow')
```

```python
# Kruskal Algorithm
mst = tree.minimum_spanning_edges(G, algorithm='kruskal', data=False)
edgelist = list(mst)
sorted(edgelist)

G2 = nx.Graph()
G2.add_edges_from(edgelist)

nx.draw(G2,  with_labels = True, node_color='yellow')
```

```python
# create random graph
n = 20
p = 0.2
g = erdos_renyi_graph(n, p)

pos = nx.spring_layout(g)
nx.draw(g, with_labels = True, edge_color='grey', node_color='#74d4e8')
```

```python
# ------ A * search ------
G2 = nx.Graph()
G2.add_edges_from([ [0,1,{'weight':2}],
                   [1,2,{'weight':12}],
                   [2,3,{'weight':1}],
                   [3,4,{'weight':15}],
                   [4,5,{'weight':1}],
                   [3,5,{'weight':8}],
                   [5,0,{'weight':10}] ])
nx.draw(G2, with_labels = True, edge_color='grey', node_color='#74d4e8')

print (nx.astar_path(G2,0,4))
print (nx.astar_path_length(G2,0,3))
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_edges_from(
    [('A', 'B'), ('A', 'C'), ('D', 'A'), ('D', 'B'), ('B', 'D'),
     ('C', 'A'), ('C', 'B'), ('C', 'B')])

nx.draw(G,  with_labels = True, node_color='yellow')

pr = nx.pagerank(G, alpha=0.4)
pr
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_edges_from(
    [('G','A'),('D','A'),('C','A'),('B','A'),('I','B'),('B','D'),('C','B'),('D','C'),('E','D'),('F','D'),('H','C')])

nx.draw(G,  with_labels = True, node_color='yellow')

hits = nx.hits(G, max_iter = 50, normalized = True)  # 8 is maximum iteration
h,a = hits
print('authority score \n')
print(a)

print('\n \n hub score \n')
print(h)
```

```python
import networkx as nx
import community as community_louvain
import matplotlib.pyplot as plt
import matplotlib.cm as cm
# define the graph
edge = [('a','b'),('a','c'),('a','d'),('a','e'),('a','f'),('b','g'),('b','h'),('b','i')]
G = nx.Graph()
G.add_edges_from(edge)

# retrun partition as a dict
partition = community_louvain.best_partition(G)
print(partition)

# visualization
pos = nx.spring_layout(G)
cmap = cm.get_cmap('cividis', max(partition.values()) + 1)
nx.draw_networkx_nodes(G, pos, partition.keys(), node_size=100,cmap=cmap, node_color=list(partition.values()))
nx.draw_networkx_edges(G, pos, alpha=0.5)
plt.show()
```
